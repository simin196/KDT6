```python
kindDF.columns=['종류','2018 수출 중량(톤)','2018 수출 금액($1000)','2019 수출 중량(톤)','2019 수출 금액($1000)','2020 수출 중량(톤)','2020 수출 금액($1000)','2021 수출 중량(톤)','2021 수출 금액($1000)']
kindDF
```

```python
kindDF.drop(index=kindDF.index[:3],inplace=True)
```

```python
kindDF=kindDF.set_index('종류')
kindDF
```

```python
ton = kindDF[['2018 수출 중량(톤)','2019 수출 중량(톤)','2020 수출 중량(톤)','2021 수출 중량(톤)']]
ton
```

```python
money= kindDF[['2018 수출 금액($1000)','2019 수출 금액($1000)','2020 수출 금액($1000)','2021 수출 금액($1000)']]
money
```

```python
# 대비 증감율(%)
idr_ton=(kindDF['2021 수출 중량(톤)']-kindDF['2020 수출 중량(톤)'])/kindDF['2020 수출 중량(톤)']*100
print(idr_ton, type(idr_ton))
```

```python
# 실수 자릿수 조정 but object이라 할 수 없어 to_numeric사용해서 실수로 변경
idr_ton1=pd.to_numeric(idr_ton, errors='coerce') # errors='coerce' 표현할수없는 부분은 NaN으로 변환해서 출력
idr_ton2= idr_ton1.round(1)
idr_ton2
```

```python
# 시리즈를 DF에 넣기
ton['20년 대비 증감율(%)'] = idr_ton2
ton
```

```python
# 전통주 : 과실주 기타주류 리큐르 약주 일반증류주 증류식소주 청주 탁주 총계
tonDF = ton.drop(['소주','맥주','브랜디','위스키','주정'])
tonDF
```

```python
idr_money=(kindDF['2021 수출 금액($1000)']-kindDF['2020 수출 금액($1000)'])/kindDF['2020 수출 금액($1000)']*100
print(idr_money, type(idr_money))
```

```python
idr_money1=pd.to_numeric(idr_money, errors='coerce')
idr_money2= idr_money1.round(1)
idr_money2
```

```python
money['20년 대비 증감율(%)'] = idr_money2
money
```

```python
moneyDF = money.drop(['소주','맥주','브랜디','위스키','주정'])
moneyDF
```
