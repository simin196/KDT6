```python
# 아시아 나라 list로 
asian_countries = [
    "Democratic People's Republic of Korea", "Japan", "China", "Russian Federation",
    "Singapore", "Cambodia", "Philippines", "Thailand", "Bangladesh", "India",
    "Pakistan", "Uzbekistan", "Saudi Arabia", "Kuwait", "Israel"]
asia_df = selected_data[selected_data['Country'].isin(asian_countries)].set_index('Country')

# 수치형 데이터만 선택하여 평균으로 결측치를 채우는 함수
def fill_with_mean(group):
    numeric_cols = group.select_dtypes(include=['number']).columns  # 수치형 컬럼만 선택
    group[numeric_cols] = group[numeric_cols].fillna(group[numeric_cols].mean())  # 수치형 컬럼의 평균으로 결측치 채우기
    return group

# 각 국가별로 결측치를 채우기
asia_df_filled = asia_df.groupby('Country',group_keys=False).apply(fill_with_mean)

# 전체 데이터의 평균으로 여전히 남아있는 결측치 채우기 (Status 컬럼 제외)
numeric_cols_global = asia_df_filled.select_dtypes(include=['number']).columns
asia_df_filled[numeric_cols_global] = asia_df_filled[numeric_cols_global].fillna(asia_df_filled[numeric_cols_global].mean())

# 결과 확인
asia_df_filled.info()
```
