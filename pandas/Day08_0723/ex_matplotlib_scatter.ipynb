### 산점도 그래프
- 두개의 컬럼(변수)의 관계를 시각화해서 표현 
- 연속형(수치형) 데이터에 사용함

```python
# 1. 모듈 로딩
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 2. 데이터 준비
F_data=r'C:\Users\KDP-23\Desktop\KDT\pandas\Day04_0716\auto_mpg.csv'
```

```python
# 3. CSV => DF으로 저장
dataDF=pd.read_csv(F_data)
```

```python
dataDF
```

```python
# - 컬럼별 타입 확인
dataDF.dtypes
```

```python
# 4. 연비와 관련이 높은 컬럼(변수) 확인
# 4-1. 상관계수 확인
corrSR= dataDF.corr(numeric_only=True)['mpg'] # >>> 숫자만 보겠다는 매개변수
round(corrSR['weight'], 2), corrSR
```

```python
# 4-2. 시각화 # 근데 뭐가 들어가야되는건지 우째암??
plt.scatter(dataDF.mpg, dataDF.weight, label=f" corr : {round(corrSR['weight'], 2)}")
plt.xlabel('mpg')
plt.ylabel('weight')
plt.legend()
plt.show()
```

```python
# 4-3. mpg 컬럼과 다른 컬럼과의 관계 시각화

# 그래프 담을 창 생성
fig= plt.figure(figsize=(15,10))

# 그래프 담을 공간 생성 axis
axs = fig.subplots(2,3,sharex=True)
for row in range(2):
    for col in range(3):
        col_name=corrSR.index[row*3+col+1]  
        axs[row, col].scatter(dataDF['mpg'], dataDF[col_name],
                              label=f" corr : {round(corrSR[col_name], 2)}")
        axs[row, col].set_title(f'mpg-{col_name}')
fig.suptitle("[mpg&colums]",fontsize=24)
plt.show()
```
