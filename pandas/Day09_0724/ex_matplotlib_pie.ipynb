### 파이 그래프
- 컬럼(변수)의 데이터의 분포 비율을 시각화

```python
# 1. 모듈 로딩
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 2. 데이터 준비
F_data=r'C:\Users\KDP-23\Desktop\KDT\pandas\Day04_0716\auto_mpg.csv'
```

```python
# 3. CSV => DF으로 저장
dataDF=pd.read_csv(F_data)
```

```python
dataDF
```

```python
# - 컬럼별 타입 확인
dataDF.dtypes
```

```python
# 4. 특정 컬럼의 데이터 비율
# - 자동차 생산 국가에 대한 비율
dataDF['origin'].astype('category')
```

```python
orginSR= dataDF['origin'].value_counts()
x=orginSR.index.to_list() # 시리즈
h=orginSR.values.tolist() # 어레이
x, h
```

```python
# - 자동차 생산국가에 대한 비율
round(dataDF['origin'].value_counts()/dataDF['origin'].count(), 2)
```

```python
# - 자동차 생산국가에 대한 비율 시각화 - bar그래프
orginBar=plt.bar(x, h, color=['blue','yellow','green'], edgecolor=['blue']) # x축에 뭐 들갈지, y축에 뭐 들갈지, 기타옵션


# 막대에 라벨을 설정
plt.bar_label(orginBar, h, label_type='center')

# x축 눈금 설정
plt.xticks([1,2,3], ['USA','EU','JA'], rotation=45, fontsize=15) # [] : 빈 리스트는 ticks이 사라진다 , 첫[]는 눈금의 위치값, 두번째[]는 넣을 값 
plt.show()
```

```python
# - 자동차 생산국가에 대한 비율 시각화 - pie 그래프
plt.pie(h,labels=['USA','EU','JA'], autopct='%.1f%%') # >> 원 그래프 
plt.show()
```

```python
# - 자동차 생산국가에 대한 비율 시각화 - pie 그래프
plt.pie(h,labels=['USA','EU','JA'], autopct='%.1f%%', counterclock=False, startangle=90) 
plt.show()
```

```python
# - 자동차 생산국가에 대한 비율 시각화 - pie 그래프
plt.pie(h,labels=['USA','EU','JA'], autopct='%.1f%%', counterclock=False, startangle=90,
        explode=[0, 0, 0.2])
plt.show() 
```

```python
# - 자동차 생산국가에 대한 비율 시각화 - pie 그래프
plt.pie(h,labels=['USA','EU','JA'], autopct='%.1f%%', counterclock=False, startangle=90,
        wedgeprops={'linewidth':10, 'width':0.5, 'edgecolor':'w'})
plt.show() 
```
