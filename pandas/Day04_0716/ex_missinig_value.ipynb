```python
# 1. 모듈 로딩
import pandas as pd
```

```python
# 2. 데이터 준비
file_name='data.csv'
```

```python
# 3. csv를 DataFrame으로
dataDF=pd.read_csv(file_name)
dataDF
```

```python
# 4. 자료 확인
```

```python
# - 실제 데이터와 컬럼별 타입 체크 >
dataDF.head()
```

```python
# 5. 데이터 전처리
# 5-1 결측치 체크 --> isnull( ), isna( )메서드
dataDF.isna() # True False로 나타내고
```

```python
dataDF.isna().sum() # NaN 몇개있는지
```

```python
# 5.2 결측치 처리 => 방법 1) 삭제 dropna()메서드
# - 기본값 : 행방향, 행 안에 1개의 컬럼이라도 Na면 삭제.
dataDF.dropna()
```

```python
# - 설정값 : 1열방향, 열 안에 1개의 원소라도 Na면 삭제.
dataDF.dropna(axis=1)
```

```python
# - 설정값 : 모든 원소가 NaN일때 삭제 how='all'
dataDF.dropna(axis=0, how='all')
```

```python
# - 설정값 : 정상데이터가 지정된 개수 이하이면 삭제 : thresh=정수
dataDF.dropna(axis=0, thresh=4)
```

```python
dataDF.dropna(axis=0, thresh=3)
```

```python
# - 설정값 : 특정 컬럼만 NaN여부 검사 진행 후 결정
# - 나이와 이름을 분석하고 싶음 => 중요한 컬럼 age
dataDF.dropna(axis=0, subset=['age','name'])
```

```python
# 5.2 결측치 처리 => 방법 2) 대체/치환 fillna()메서드
# - 대체값 : 원하는 값 또는 컬럼의 평균값/최빈값 등을 사용 가능
#           이전의 값 또는 이후의 값으로 대체
dataDF.fillna(0)
```

```python
dataDF.fillna({'name':'unknown'},{'age':10})
```

```python
# dataDF.fillna(method='bfill') >> 미래에 제거되는 매개변수 method
dataDF.bfill()
```

```python
# - 나이의 결측치는 나이 평균으로 채우기
dataDF.fillna({'age':dataDF['age'].mean()}, inplace=True)
```

```python
# - 이름의 결측치는 가장 많이 사용된 이름으로 채우기
dataDF.fillna({'name':dataDF['name'].mode()}, inplace=True)
```
