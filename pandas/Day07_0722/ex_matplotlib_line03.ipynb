### Matplotlib 활용 데이터 시각화
- x축과 y축 다양한 처리

```python
# 1. 모듈 로딩
from matplotlib import pyplot as plt 
# 패키지 안에 여러 모듈 중 하나 선택해서 로딩하는 경우 from을 사용
# from 패키지명 import 모듈명 as 별명
# 이때까지 import matplotlib.pyplot as plt 써왔는데 이 방법도 있다 정도 알고있음 좋을듯
```

```python
# 2. 그래프 그리기 ------------- 기본
y_data=[1,3,5,7,9]
x_data=['2024-01-01','2024-02-01','2024-03-01','2024-04-01','2024-05-01']
plt.plot(x_data, y_data, marker='o')
plt.show()
```

```python
# 2. 그래프 그리기 ------------- x축과 y출 범위 설정
y_data=[1,3,5,7,9]
x_data=['2024-01-01','2024-02-01','2024-03-01','2024-04-01','2024-05-01']
plt.plot(x_data,y_data, marker='o')
plt.ylim(bottom=0, top=15)
# plt.xlim(left='2023-12-01', right='2024-06-01')>> 범위 설정
plt.show()
```

```python
# 2. 그래프 그리기 ------------- x축과 y축 범위 설정
y_data=[1,3,5,7,9]
x_data=['2024-01-01','2024-02-01','2024-03-01','2024-04-01','2024-05-01']
plt.plot(x_data,y_data, marker='o')
plt.ylim(bottom=0, top=15)
plt.xticks(rotation=45) #>>>>???? 이건 뭐레??
plt.show()
```

```python
# 2. 그래프 그리기 ------------- 그래프 출력 창 즉, figure 크기 설정
y_data=[1,3,5,7,9]
x_data=['2024-01-01','2024-02-01','2024-03-01','2024-04-01','2024-05-01']

plt.figure(figsize=[14,10])  # 원하는 창의 크기 설정 
plt.plot(x_data,y_data, marker='o')
plt.ylim(bottom=0, top=15)
plt.show()
```

```python
# [2] 그래프 그리기  -  그래프 출력 창 즉, figure 크기 설정
# 위 그래프와 비교(크기비교)
plt.figure(figsize=[10,5])     #>>> 원하는 창의 크기 설정 
plt.plot(x_data, y_data, marker='o')
plt.ylim(bottom=0, top=15)
plt.show()
```

```python
# 2. 그래프 그리기 ------------- 그래프 출력 창 즉, figure를 자동 맞춤
y_data=[1,3,5,7,9]
x_data=['2024-01-01','2024-02-01','2024-03-01','2024-04-01','2024-05-01']

plt.plot(x_data, y_data, marker='o')
plt.ylim(bottom=0, top=10)
plt.tight_layout()          # 모든 내용 출력될 수 있도록 자동 맞춤 설정
plt.show()
```

```python
# 3. 그래프 그리기 ------------- 눈금(stcks) 설정
y_data=[1,3,5,7,9]
x_data=['2024-01-01','2024-02-01','2024-03-01','2024-04-01','2024-05-01']

plt.plot(y_data, marker='o')
plt.ylim(bottom=0, top=15)
plt.xlim(left=0, right=7)
plt.xticks([0,1,2,3,4,5,6,7]) # 리미트에 맞춰서 줘야 0이 0에서 시작한다. 아니면 앞에 튀 나온다
plt.show()
```

```python
y_data=[1,3,5,7,9]
x_data=['2024-01-01','2024-02-01','2024-03-01','2024-04-01','2024-05-01']

plt.plot(y_data, marker='o')
plt.ylim(bottom=-1, top=15) # '-'를 넣어줄수도 있다.
plt.xlim(left=-1, right=6)
plt.xticks([0,2,4,6],['A','B','C','D'], fontsize=22, fontweight='bold')
plt.show()
```

```python
y_data=[1,3,5,7,9]
x_data=['2024-01-01','2024-02-01','2024-03-01','2024-04-01','2024-05-01']

plt.plot(y_data, marker='o')
plt.ylim(bottom=-1, top=15) # '-'를 넣어줄수도 있다.
plt.xlim(left=0, right=6)

font_config={'fontsize':22, 'fontweight':'bold'}
plt.xticks([0,2,4,6],['A','B','C','D'], **font_config) #<<< 가변인자

plt.show()
```
