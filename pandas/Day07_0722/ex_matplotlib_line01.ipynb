#### Matplotlib를 활용한 시각화
- Matplotlib.pyplot 모듈을 사용
- 다양한 그래프 관련 함수들 & 클래스들 존재

```python
# 1. 모듈 로딩
# 패키지명. 모듈명
import matplotlib.pyplot as plt
```

```python
# 2. 그래프 그리기
# 모듈명.함수 # plot은 선그래프 - 2차원 평면 # 객체를 변수로 저장하면 나중에 데이터확인이나 다른 정보 확인할 때 편하다.
# 이때 y축이 기본값
lineObject=plt.plot([1, 2, 3, 4])
plt.xlabel('DATA') # 축에 이름 붙힐수있다.
plt.ylabel('result')
plt.show()
```

```python
plt.plot([10,20,30],[5,10,15])
plt.show()
```

```python
# 그래프 그리기 - Line(선) 그래프
# plot(x축 데이터, y축 데이터, fmf 포멧스트림)
# fmf 포멧스트링 : 그래프의 색상, 두께, 스타일 등등 지원
plt.plot([10,20,30],[5,10,15], 'r--')
plt.show()
```

```python
# 그래프 그리기 - Line(선) 그래프
# plot(x축 데이터, y축 데이터, fmf 포멧스트림)
# fmf 포멧스트링 : 그래프의 색상, 두께, 스타일 등등 지원
plt.plot([10,20,30],[5,10,15], 'ro--') # '색깔, 표시모양, 선 모양'
plt.show()
```

```python
plt.plot([10,20,30],[5,10,15], 'r^--') # 위로가는 삼각형
plt.show()
```

```python
plt.plot([10,20,30],[5,10,15], 'b>--') # 원하는 모양으로 대입할 수 있다.
plt.show()
```

```python
plt.plot([10,20,30],[5,10,15], 'k>--') 
plt.show()
```

```python
plt.plot([10,20,30],[5,10,15], 'b+')  # 선 안그림, 선모양을 안그리고 값이 나오는 spot에 생긴다.
plt.show()
```

```python
# 그래프 그리기 - Line(선) 그래프
# plot(x축 데이터, y축 데이터, line2D속성으로 설정)
# fmf 포멧스트링 : line2D속성으로 설정
# - color="red" , marker='o' , linestyle='dashed', 
# - markersize=10(마크 사이즈), linewidth=4(선 사이즈)
# - markeredgecolor => mec='blue'(마크 테두리 색깔)
# - markeredgewidth => mew=5(마크 테두리 두께)
plt.plot([10,20,30],[5,10,15],
        color="red" , marker='o' , linestyle='dashed', 
        markersize=10, linewidth=4,  mec='blue', mew=5) 
plt.show()
```

```python
# 그래프 그리기 - Line(선) 그래프
# plt.plot([10,20,30],[15,25,35], 'bo--')
# plt.plot([10,20,30],[18,29,25], 'ro--')
# plt.plot([10,20,30],[25,25,25], 'k^-') 한번에
plt.plot([10,20,30],[15,25,35], 'bo--',[10,20,30],[18,29,25], 'ro--', [10,20,30],[25,25,25], 'k^-')
plt.show()
```

```python
# 3. 그래프의 부가적인 기능 추가
# - 그래프 제목
# - x축, y축 라벨
# - 여러개 그래프 경우 범례 ==> 그래프의 label속성값을 범례에서 사용함
plt.plot([10,20,30],[15,25,20],'bo--', label="V")
plt.plot([10,20,30],[5,30,28],'ro--', label="2")
plt.title('[MPLIB LIVE2D]')
plt.xlabel('Data') 
plt.ylabel('Result')
plt.legend()
plt.show()
```

```python
# 3. 그래프의 부가적인 기능 추가
# - 그래프 제목
# - x축, y축 라벨
# - 여러개 그래프 경우 범례 ==> 그래프의 label속성값을 범례에서 사용함
# -                           loc속성으로 위치 지정 가능 [기본값:best]
plt.plot([10,20,30],[15,25,20],'bo--', label="V")
plt.plot([10,20,30],[5,30,28],'ro--', label="2")
plt.title('[MPLIB LIVE2D]')
plt.xlabel('Data') 
plt.ylabel('Result')
plt.legend(loc='upper right') #기본값:best 최대한 그래프가 잘보이는 위치에 있는다 but 이렇게 위치도 지정 가능하다.
plt.show()
```
