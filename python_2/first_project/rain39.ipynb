```python
## 이거 3월부터 9월까지의 데이터

import pandas as pd 
from tabulate import tabulate
import seaborn as sns
import koreanize_matplotlib
import matplotlib.pyplot as plt
```

```python
rain=pd.read_excel('new_rain_from3to9.xlsx')
print(tabulate(rain.head(),headers='keys',tablefmt='psql'))
```

```python
def renew(data):
    data.drop(['지점','지점명'], axis=1, inplace=True)
    data.columns=['date','temp','rain','wind','humidity']
    data['date']= pd.to_datetime(data['date'].dt.date)
    data.ffill(inplace=True)
    data['rain'] = data['rain'].replace(0,0.01)
    data['rain'] = data['rain'].fillna(0.01)
    print(data['rain'].value_counts())
    print(data.columns)
    print()
    print(data.isna().sum())
    print()
    print(tabulate(data.head(), headers='keys', tablefmt='pretty'))
```

```python
renew(rain)
```
