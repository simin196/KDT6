```python
# 데이터 파일 읽기
with open(MNIST_PATH + filelist[0], mode='rb') as f:
    
    # # 파일의 헤드 정보 읽기 1번
    # data = f.read(8)
    # print(struct.unpack('>II' , data)) # 탑 엔디언 # 앞에 8자리는 은 매직 코드, 데이터 개수
    # print(struct.unpack('<II' , data)) # 리틀 엔디언
    
    # 2번 
    _, cnt = struct.unpack('>II' , f.read(8) ) # 매직 코드, 데이터 개수
    rows, cols = struct.unpack('>II' , f.read(8)) # 행, 열

    # 우리가 가지고올 이미지 raw 데이터
    pixels = f.read(rows*cols) # >> 실제 픽셀값
    print(pixels) 
```

```python
# 데이터 파일 읽기
with open(MNIST_PATH + filelist[0], mode='rb') as data_f: # 데이터 파일
    with open(MNIST_PATH + filelist[1], mode='rb') as label_f: # 라벨 파일
    
        # 파일의 헤드 정보 읽기 
        _, img_cnt = struct.unpack('>II' , data_f.read(8)) 
        rows, cols = struct.unpack('>II' , data_f.read(8)) 

        _, _ = struct.unpack('>II' , label_f.read(8) ) 

        for n in range(1):
            # 이미지 raw 데이터
            pixels = data_f.read(rows*cols) 
            sdata = list(map(lambda n: str(n), pixels))
            print(sdata)

            # 이미지 라벨
            label = struct.unpack('B', label_f.read(1))[0]
            print(label)

            # 피쳐와 라벨 연결
            sdata.append (str(label))
            print(sdata)

            # ','.join(sdata)
            print(','.join(sdata))
```

```python
# 데이터 파일 읽어서 데이터셋 파일 생성
def makeDataFile(imgfile, labelfile, datasetfile, datalength=None): # datalength=None 데이터 개수 조절가능
    with open(imgfile, mode='rb') as data_f: # 데이터 파일
        with open(labelfile, mode='rb') as label_f: # 라벨 파일
            with open(datasetfile, mode='w') as dataset_f: # 저장할 폴더 파일


                # 파일의 헤드 정보 읽기 
                _, img_cnt = struct.unpack('>II' , data_f.read(8)) 
                rows, cols = struct.unpack('>II' , data_f.read(8)) 

                struct.unpack('>II' , label_f.read(8) ) 

                for n in range(img_cnt):
                    # 이미지 raw 데이터
                    pixels = data_f.read(rows*cols) 
                    sdata = list(map(lambda n: str(n), pixels))
              
                    # 이미지 라벨
                    label = struct.unpack('B', label_f.read(1))[0]
               

                    # 피쳐와 라벨 연결
                    sdata.append (str(label)) # 뒤에다가
                    # sdata.insert(0, str(label)) # 앞에다가 insert원하는 위치에 삽입하는거라 인덱스 줘야함
                   

                    # 데이터 파일에 기록
                    img_label = ','.join(sdata) #디버깅
                    dataset_f.write(img_label+'\n')
                  
```

```python
DS_PATH = '../data/'
MNIST_PATH = '../data/MNIST/raw/'

Train_File = DS_PATH + 'mnist_train.csv'
Test_File = DS_PATH + 'mnist_test.csv'
```
