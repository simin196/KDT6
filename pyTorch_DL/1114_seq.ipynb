```python
## ANN 모델  입 : 출 
# 입력층 :   3개  10개  >> 10개 뉴런 ==> AF : ReLu
# 은닉층 :  10개  30개  >> 30개 뉴런 ==> AF : ReLu
# 출력층 :  30개   1개  >>  1개 뉴런 ==> AF : 회귀 linear >> pytorch X
mModel = nn.Sequential(
    nn.Linear(3,10),
    nn.ReLU(),
    nn.Linear(10,30),
    nn.ReLU(),
    nn.Linear(30,1)
)
```

```python
# 임시 데이터 ==> 2차원 (n,3) : n개 데이터  >> 왜 3이냐 >> 피쳐 3개라서
data = torch.FloatTensor([[10,20,30],[11,22,33]])
data.shape

target = torch.FloatTensor([[9,12]])
```

```python
# 순방향 학습 ==> 내부적으로 forward() 메서드 호출
# [전달] [입력층] f1w1 + f2w2 + f3w3 + b >> ReLU() >> output 10개
#        [은닉층] >> o1w1 + o2w2 + o3w3 + ... + o10w10 + b >> ReLU() >> output 30개
#        [출력층] >> o1w1 + o2w2 + o3w3 + ... + o10w10 + b >> output 1개
mModel(data)
```
