```python
# 피클 파일이 있는 폴더 경로
folder_path = './pkl'  
title = []
text = []
opinion = []
director = []

# 'list'가 올바르게 파이썬의 기본 타입인지 확인
# print(type(list))  # 기본적으로는 <class 'list'>가 출력되어야 합니다. >>> 이게 뭔말이지??
```

```python
# 폴더 경로가 존재하는지 확인
if not os.path.exists(folder_path):
    print(f"폴더 경로가 존재하지 않습니다: {folder_path}")

else:
    # 폴더 내 모든 파일 탐색
    for filename in os.listdir(folder_path):
        
        # 'title'이 포함된 파일 처리
        if 'title' in filename and (filename.endswith('.pickle') or filename.endswith('.pkl')):
            file_path = os.path.join(folder_path, filename)  # 파일 경로 생성
            with open(file_path, 'rb') as f:
                data = pickle.load(f)
                title.extend(data)


        # 'text'가 포함된 파일 처리
        elif 'text' in filename and (filename.endswith('.pickle') or filename.endswith('.pkl')):
            file_path = os.path.join(folder_path, filename)
            with open(file_path, 'rb') as f:
                data = pickle.load(f)
                text.extend(data)


        # 'opinion'이 포함된 파일 처리
        elif 'opinion' in filename and (filename.endswith('.pickle') or filename.endswith('.pkl')):
            file_path = os.path.join(folder_path, filename)
            with open(file_path, 'rb') as f:
                data = pickle.load(f)
                opinion.extend(data)


        # 'director'가 포함된 파일 처리
        elif 'director' in filename and (filename.endswith('.pickle') or filename.endswith('.pkl')):
            file_path = os.path.join(folder_path, filename)
            with open(file_path, 'rb') as f:
                data = pickle.load(f)
                director.extend(data)


print("title:", len(title))
print("text:", len(text))
print("opinion:", len(opinion))
print("director:", len(director))
```

```python
opinion_1 = []

for txt in opinion:
    result = re.sub(r'\s+', ' ', txt)
    L = []
    
    for t in result.split('|'):
        L.append(t.strip())

    opinion_1.append(L)

opinion_1
```

```python
def categorize_elements(opin):

    # 카테고리 리스트 초기화
    genre, kind, time, old, long, country = '', '', '', '', '', ''
    
    for item in opin:

        # 숫자가 있는 경우 '시간'로 분류
        if re.search(r'\d+분|\d+초', item):  
            time = item

         # '관람'라는 단어가 포함된 경우 '등급'으로 분류
        elif ('관람가' in item) or ('관람불가' in item):
            old = item

        # 장르분류
        elif item in ['드라마', '코미디', '액션', '멜로/로맨스', '스릴러', '미스터리', '공포(호러)',
                    '어드벤처', '범죄', '가족', '판타지', 'SF', '서부극(웨스턴)', '사극', '애니메이션',
                    '다큐', '전쟁', '뮤지컬', '성인물(에로)', '공연', '기타']:
            genre = item

        # 영화 분류
        elif item in ['장편' ,'단편', '옵니버스', '온라인전용']: 
            long = item

        # 영화 구분 분류
        elif item in ['일반영화', '예술영화', '독립영화']: 
            kind = item

        # 국가 분류
        else:
            country = item

    return [genre, kind, time, old, long, country]

# 새로운 리스트 생성
opinion_sorted = []
for opin in opinion_1:
    opinion_sorted.append(categorize_elements(opin))
```

```python
title_SD = []
for tit in title:
    title_SD.append([tit])

director_SD = []
for dir in director:
    director_SD.append([dir])

text_SD = []
for t in text:
    text_SD.append([t])
```

```python
for i in range(len(title)):
    title_SD[i].extend(opinion_sorted[i])
    title_SD[i].extend(director_SD[i])
    title_SD[i].extend(text_SD[i])
```
