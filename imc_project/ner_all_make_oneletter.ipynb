```python
import os
import json
```

```python
# 파일 읽기
try:
    with open('./data/tagged_more_Data.json', 'r', encoding='utf-8') as file:
        ner_all = json.load(file)
except FileNotFoundError:
    raise Exception("파일을 찾을 수 없습니다. 경로를 확인하세요.")
except json.JSONDecodeError:
    raise Exception("JSON 파일이 잘못된 형식입니다.")
```

```python
ner_all
```

```python
# 변환된 데이터 저장
processed_data = []

for entry in ner_all:
    text = "".join(entry["tokens"])  # 토큰을 결합하여 원문 생성
    tokens = []  # 새롭게 쪼갠 토큰
    ner_tags = []  # 새롭게 매핑된 NER 태그

    for token, tag in zip(entry["tokens"], entry["ner_tags"]):
        if len(token) == 1:
            tokens.append(token)
            ner_tags.append(tag)
        else:
            # 한 글자씩 분리
            for i, char in enumerate(token):
                tokens.append(char)
                # 첫 번째 글자는 1, 3, 5 태그, 나머지는 2, 4, 6 태그
                if tag in [1, 3, 5]:
                    ner_tags.append(tag if i == 0 else tag + 1)
                else:
                    ner_tags.append(tag)

    # 새 포맷으로 저장
    processed_data.append({
        "tokens": tokens,
        "ner_tags": ner_tags
    })
```

```python
# 파일 저장
for i in range(len(processed_data)):
    with open("tag_oneletter_addone.json", "a", encoding="utf-8") as f:
        f.write(f'''
{{
"tokens" : {str(processed_data[i]["tokens"]).replace("'",'"')},
"ner_tags" : {processed_data[i]["ner_tags"]}
}},''')

print('파일 저장 완료')
```
